# Questão 1

Implemente uma função load_data(seed) que:

Carregue o dataset `Fashion MNIST`
Realize a separação em treino e teste
Utilize `train_test_split` com controle de aleatoriedade
Retorne: `X_train`, `X_test`, `y_train`, `y_test`

Depois responda: 
É necessário normalizar os dados para esse tipo de modelo? Justifique.

**Solução**:

````python
# TODO: implemente load_data
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split

def load_data(seed=42):

    X, y = fetch_openml(
        "Fashion-MNIST",
        version=1,
        as_frame=False,
        return_X_y=True
    )

    y = y.astype(int)

    X_train, X_test, y_train, y_test = train_test_split(
        X, y,
        test_size=0.2,
        random_state=seed,
        stratify=y
    )

    return X_train, X_test, y_train, y_test
````

Não é necessário normalizar dados em algoritmos que de arvores de decisão, random forest e adaboost, pois eles eles realizam divisao por meio das limiares das features e nao utiliza a distancia entre pontos.

# Questão 2

Implemente as funções:

`train_random_forest(X_train, y_train, seed)`
`train_adaboost(X_train, y_train, seed)`

## Requisitos:

Utilizar os modelos do `sklearn`
Garantir reprodutibilidade com `random_state`

**Solução**:

````python
# TODO: implemente load_data

from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier

def train_random_forest(X_train, y_train, seed=42):
    model = RandomForestClassifier(random_state=seed)
    model.fit(X_train, y_train)
    return model

def train_adaboost(X_train, y_train, seed):
    model = AdaBoostClassifier(random_state=seed)
    model.fit(X_train, y_train)
    return model
````

# Questão 3

Ela deve:

# Questão 4

Implemente a função:

- `run_pipeline(model_type="rf", seed=42)`

Ela deve:
- Carregar os dados
- Treinar o modelo escolhido (`rf` ou `ab`)
- Avaliar o modelo
- Retornar a acurácia

**Solução**:

````python
# TODO: implemente load_data

def run_pipeline(model_type="rf", seed=42):
    X_train, X_test, y_train, y_test = load_data(seed)
    if(model_type == "rf"):
        model = train_random_forest(X_train, y_train, seed)
    elif(model_type == "ab"):
        model = train_adaboost(X_train,y_train, seed)
    else:
        print("Escolher ou rf ou ab")
    acc = evaluate(model, X_test, y_test)
    return acc
````

````python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = load_data(42)

for depth in range(1, 11):

    tree = DecisionTreeClassifier(max_depth=depth, random_state=42)
    tree.fit(X_train, y_train)

    train_acc = accuracy_score(y_train, tree.predict(X_train))
    test_acc = accuracy_score(y_test, tree.predict(X_test))

    print(f"depth={depth} | treino={train_acc:.4f} | teste={test_acc:.4f}")
````

**Em qual profundidade começa o overfitting?**
**Por que a árvore consegue 100% no treino quando max_depth=None?**

O overfitting começa na profundidade 4, pois a partir desse valor a acurácia no conjunto de treino continua aumentando, enquanto a acurácia no conjunto de teste deixa de melhorar ou passa a diminuir. Isso indica que o modelo está se ajustando demais aos dados de treino e perdendo capacidade de generalização.

Porque quando max depth igual NONE, a altura da árvore não tem um limite e a árvore vai ficar cada vez mais profunda até todos os nós estarem puros. Isso é overfitting, pois no treino dessa arvore ela vai ficar com 100% e no teste a sua acuracia vai cair muito.

# Questão 5

Execute o pipeline para ambos os modelos:

- Random Forest
- AdaBoost

## Apresente:
- Acurácia, Precisão, Recall e F1-Score de cada modelo

## Responda:
- Qual modelo apresentou melhor desempenho inicial?
 
 RESPOSTA: Random Forest que teve valores de acuracia e f1 score maiores se comparado ao adaboost
 

**Solução**:

````python
# TODO: implemente load_data
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

X_train, X_test, Y_train, Y_test = load_data(42)

rf_model = train_random_forest(X_train, Y_train, 42)
ab_model = train_adaboost(X_train, Y_train, 42)

Y_predRF = rf_model.predict(X_test)
Y_predAB = ab_model.predict(X_test)

accRF = accuracy_score(Y_test, Y_predRF)
accAB = accuracy_score(Y_test, Y_predAB)

precisionRF = precision_score(Y_test, Y_predRF, average = "macro")
precisionAB = precision_score(Y_test, Y_predAB, average = "macro")

recallRF = recall_score(Y_test, Y_predRF, average = "macro") 
recallAB = recall_score(Y_test, Y_predAB, average = "macro") 

f1scoreRF = f1_score(Y_test, Y_predRF, average="macro")
f1scoreAB = f1_score(Y_test, Y_predAB, average="macro")

print("Random Forest")
print("Accuracy:", accRF)
print("Precision:", precisionRF)
print("Recall:", recallRF)
print("F1-score:", f1scoreRF)

print("AdaBoost")
print("Accuracy:", accAB)
print("Precision:", precisionAB)
print("Recall:", recallAB)
print("F1-score:", f1scoreAB)
````

# Questão 8

Varie pelo menos um hiperparâmetro em cada modelo:

- Random Forest: `n_estimators`
- AdaBoost: `n_estimators`

## Analise:
- O desempenho muda significativamente?

## Responda:
- Qual modelo é mais sensível a mudanças?

RESPOSTA: adaboosting

````python
# TODO: implemente load_data
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import AdaBoostClassifier

X_train, X_test, y_train, y_test = load_data(42)

for n in [10, 50, 100, 200]:

    model = RandomForestClassifier(
        n_estimators=n,
        random_state=42
    )

    model.fit(X_train, y_train)

    acc = accuracy_score(y_test, model.predict(X_test))

    print("RF n_estimators =", n, ":", acc)
    

for n in [10, 50, 100, 200]:

    model = AdaBoostClassifier(
        n_estimators=n,
        random_state=42
    )

    model.fit(X_train, y_train)

    acc = accuracy_score(y_test, model.predict(X_test))

    print("AB n_estimators =", n, ":", acc)
````

1) Não. A acurácia é uma métrica importante, mas não é suficiente para avaliar completamente o desempenho de um modelo, especialmente em problemas de classificação multiclasse. Ela indica apenas a proporção total de acertos, mas não mostra como o modelo se comporta em relação aos erros específicos entre as classes. Por isso, métricas como precisão, recall e F1-score são importantes, pois avaliam diferentes aspectos do desempenho do modelo. No experimento realizado, utilizamos essas métricas adicionais para obter uma avaliação mais completa dos modelos Random Forest e AdaBoost.

2) Para garantir que os resultados não ocorreram por acaso, utilizamos o parâmetro random_state tanto na separação dos dados quanto no treinamento dos modelos. Isso permite controlar a aleatoriedade do experimento e garantir que os resultados possam ser reproduzidos. Além disso, executamos o pipeline com diferentes valores de seed e observamos que os resultados permaneceram semelhantes, indicando que o desempenho dos modelos é consistente e não depende apenas de uma divisão específica dos dados.

3) Um possível problema metodológico é o uso de apenas uma divisão treino/teste para avaliar o desempenho dos modelos. Idealmente, seria mais adequado utilizar técnicas como validação cruzada (cross-validation) para obter uma estimativa mais confiável da performance. Outro problema é a ausência de ajuste de hiperparâmetros dos modelos. Os algoritmos foram utilizados com configurações padrão, o que pode não representar o melhor desempenho possível para o dataset Fashion MNIST.

4) O pipeline implementado é confiável para uma avaliação inicial dos modelos, pois organiza corretamente as etapas de carregamento dos dados, separação treino/teste, treinamento e avaliação, além de controlar a aleatoriedade com o uso de seeds. No entanto, para uma análise mais robusta, seria necessário incluir validação cruzada, ajuste de hiperparâmetros e testes adicionais com diferentes configurações. Dessa forma, o pipeline atual é adequado para experimentos iniciais, mas pode ser aprimorado para aplicações mais avançadas.
